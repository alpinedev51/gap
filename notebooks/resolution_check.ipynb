```python
from PIL import Image
import os
import numpy as np

# Path to your extracted images
data_dir = "../oxford-iiit-pet/images"
image_files = [f for f in os.listdir(data_dir) if f.endswith(".jpg")]

dims = []
for f in image_files:  # Check first 100 for a quick sample
    with Image.open(os.path.join(data_dir, f)) as img:
        dims.append(img.size)

avg_dims = np.mean(dims, axis=0)
print(f"Average Resolution: {avg_dims[0]:.1f}x{avg_dims[1]:.1f}")
print(f"Max Resolution: {np.max(dims, axis=0)}")
print(f"Min Resolution: {np.min(dims, axis=0)}")
```
